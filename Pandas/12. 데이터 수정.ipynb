# 데이터 수정

```python
import pandas as pd
df =pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'})
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
```

```python
df
```

```python
df['SW특기']=df['SW특기'].str.lower()
```

```python
df
```

```python
df['학교']=df['학교']+'등학교'
df
```

## Column 추가

```python
df['총합']=df['국어']+df['영어']
```

```python
df
```

```python
df['결과']='Fail'
```

```python
df
```

```python
df.loc[df['총합']>100, '결과']='pass'  # 총합이 100보다 큰 데이터에 대해서 결과를 pass로
```

```python
df
```
